```python
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from keras.utils import plot_model
```

```python
model = load_model('FaceNet_Model/facenet_keras.h5')
model.load_weights('FaceNet_Model/facenet_keras_weights.h5')
print(model.summary())
plot_model(model, to_file='image_data/Images_for_ReadMe/FaceNet_Model_Structure.png', show_shapes=True, show_layer_names=True)
```
