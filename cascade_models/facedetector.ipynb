```python
# check opencv version
import cv2
from cv2 import CascadeClassifier
from cv2 import imread
from cv2 import imshow
from cv2 import waitKey
from cv2 import destroyAllWindows
from cv2 import CascadeClassifier
from cv2 import rectangle

import matplotlib.patches as patches
import matplotlib.pyplot as plt
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook

import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
pixels = imread('fa.jpg')
```

```python
#cv2 displays images as BGR not RGB which is required for plt

original_image = cv2.cvtColor(pixels, cv2.COLOR_BGR2RGB)
plt.imshow(original_image)
plt.title('RGB Picture')
plt.show()
```

```python
#Display cv image unedited
plt.imshow(pixels)
plt.title('Original Picture')
plt.show()
```

```python
# load the pre-trained model
classifier = CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# perform face detection

#detectMultiScale accepts scaleFactor and minNeighbors
bboxes = classifier.detectMultiScale(pixels, 1.06, 8)
```

```python
dpi = 80
height, width, depth = pixels.shape
figsize = width / float(dpi), height / float(dpi)
```

```python
# Create figure and axes
fig = plt.figure(figsize=figsize)
ax = fig.add_axes([0, 0, 1, 1])

# Hide spines, ticks, etc.
ax.axis('off')

# Display the image
ax.imshow(pixels)

# print bounding box for each detected face
for box in bboxes:
# extract
    x, y, width, height = box
    x2, y2 = x + width, y + height
    # draw a rectangle over the pixels
    #rectangle(pixels, (x, y), (x2, y2), (0,0,255), 1)
    rect = patches.Rectangle((x,y),x2/4,y2/1.5,linewidth=1,edgecolor='r',facecolor='none')
    # Add the patch to the Axes
    ax.add_patch(rect)


    
    
plt.show()    
```
